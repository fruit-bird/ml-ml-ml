```python
from sklearn.compose import ColumnTransformer

USELESS_ATTRIBUTES = ['Date_of_Surgery', 'Date_of_Last_Visit', 'ER status', 'PR status']
breast_cancer_columns = breast_cancer.drop(USELESS_ATTRIBUTES, axis=1)

TARGET_ATTRIBUTE = 'Patient_Status'
NUM_ATTRIBUTES = ['Age', 'Protein1', 'Protein2', 'Protein3', 'Protein4']
CAT_ATTRIBUTES = ['Gender', 'Tumour_Stage', 'Surgery_type', 'Histology', 'HER2 status']

full_pipeline = ColumnTransformer([
    ('num', numeric_pipeline, NUM_ATTRIBUTES),
    ('cat', OneHotEncoder(), CAT_ATTRIBUTES),
])

prepared_breast_cancer_X = full_pipeline.fit_transform(breast_cancer_columns.drop(TARGET_ATTRIBUTE, axis=1))
pd.DataFrame(prepared_breast_cancer_X)
```

## Models

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    prepared_breast_cancer_X,
    breast_cancer[TARGET_ATTRIBUTE],
    test_size=0.2,
)
```

```python
from sklearn.ensemble import RandomForestClassifier

forest_clf = RandomForestClassifier()
forest_clf.fit(X_train, y_train)
```

```python
predictions = forest_clf.predict(X_test)
predictions[10:30], np.array(y_test[10:30])
```
